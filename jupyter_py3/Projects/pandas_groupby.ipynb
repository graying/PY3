```python
import pandas as pd
import numpy as np
cols = ['one','two','three','four','five','six','seven','eight','nine','ten']
df = pd.DataFrame({'aaa': np.random.randint(4, size=20),
                   'bbb': np.random.random(20)*100},
                    );df
```

```python
%matplotlib inline
df.groupby('aaa')['bbb'].sum().plot(kind='bar')
```

### 从每个aaa的分类中，选出bbb最大的值 

```python
gbaaa = df.groupby('aaa')
for gb in np.sort(df.aaa.unique()):
    bbbmax = gbaaa.get_group(gb).bbb.max()
    print ('group:'+str(gb)+' max is ',bbbmax)
```
