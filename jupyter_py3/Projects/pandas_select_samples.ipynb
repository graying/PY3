#### use 'where' function to seleted 
pd.where(mask, value) please be ware the value will be given to 'False' postion, the True will remain their value.

```python
pd1 = pd.DataFrame({'aaa':[0,1,2,3,4,5],
                    'bbb':[3,4,5,6,7,8],
                    'ccc':[4,5,6,7,8,9]}); pd1
mask1=pd.DataFrame({'aaa':[True]*6,
                    'bbb':[False]*6,
                    'ccc':[True, False]*3})
pd1.where(mask1, 'FALSE')
```

### user numpy's where function to give value in if...then...else style

```python
pd2 = pd.DataFrame({'aaa':[0,1,2,3,4,5],
                    'bbb':[3,4,5,6,7,8],
                    'ccc':[4,5,6,7,8,9]})
import numpy as np
pd2['over3']=np.where(pd2.aaa>3,True,False);pd2
```

### splitting the DataFrame

```python
over3 = pd2[pd2.over3==True]; over3
```

```python
low3 = pd2[pd2.over3==False]; low3
```

### argsort 可以排序调用他的Serials，并返回一个serials用于loc

```python
pd3 = pd.DataFrame({'AAA':np.arange(10),
                    'BBB':np.random.random(10)*1000
});pd3
```

```python
pd3.loc[(pd3.BBB).argsort()]
```

```python
crit1 = pd3.BBB > 300
crit2 = pd3.BBB < 600
''' use ~ to inverse the mask'''
pd3[~crit1]
```
