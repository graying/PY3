```python
%matplotlib notebook
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = np.loadtxt("populations.txt") # 从文件中读取数据
```

```python
year, hares, lynxes, carrots = data.T # 技巧，将列转为行，然后赋给各个变量
```

```python
data.T
```

```python
plt.axes([0.2,0.1,0.5,0.8])
plt.plot(year,hares,year,lynxes,year,carrots)
plt.legend(('Hare',"Lynx", "Carrot"), loc=(1.05, 0.5))
```

```python
populations = data[:, 1:]
```

```python
populations.mean(axis=0) #平均数
```

```python
populations.std(axis=0) #标准差
```

```python
populations
```

```python
np.max(populations, axis=1) #Populations 数据中，每一行最大的值
```

```python
np.argmax(populations, axis=1) #Populations 数据中，每一行最大的值的索引（下标）
```
